```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 500)

# Definición de todos los polinomios de Legendre
P0 = np.ones_like(x)                         # Grado 0
P1 = x                                       # Grado 1
P2 = (3/2)*x**2 - (1/2)                     # Grado 2
P3 = (1/2)*(5*x**3 - 3*x)                   # Grado 3
P4 = (35/8)*x**4 - (15/4)*x**2 + 3/8        # Grado 4
P5 = (1/8)*(63*x**5 - 70*x**3 + 15*x)       # Grado 5
P6 = (1/16)*(231*x**6 - 315*x**4 + 105*x**2 - 5)  # Grado 6
P7 = (1/16)*(429*x**7 - 693*x**5 + 315*x**3 - 35*x)  # Grado 7

# Configuración del gráfico
plt.figure(figsize=(12, 8))

# Graficar polinomios impares (líneas continuas)
plt.figure(figsize=(10, 6))
plt.plot(x, P1, 'b-', label='$P_1(t)$')
plt.plot(x, P3, 'r-', label='$P_3(t)$')
plt.plot(x, P5, 'm-', label='$P_5(t)$')
plt.plot(x, P7, 'c-', label='$P_7(t)$')


plt.legend(loc='lower right', fontsize=19)
plt.xlabel('t', fontsize=20)
plt.ylabel('$P_n(t)$',fontsize=20)
plt.title('Polinomios de Legendre de grado impar', fontsize=22)
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(-1.2, 1.2)  # Ajuste de rango vertical para mejor visualización
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.savefig('legendreimpar.png', dpi=300, bbox_inches='tight')

plt.show()
```
